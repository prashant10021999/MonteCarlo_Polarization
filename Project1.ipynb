```python
import numpy as np

def read_dat_file(file_path):
    """Reads a .dat file and returns the data as a numpy array."""
    return np.loadtxt(file_path)

# Read all the required .dat files
HI = read_dat_file('outHI.dat')
HQ = read_dat_file('outHQ.dat')
HU = read_dat_file('outHU.dat')
HV = read_dat_file('outHV.dat')

PI = read_dat_file('outPI.dat')
PQ = read_dat_file('outPQ.dat')
PU = read_dat_file('outPU.dat')
PV = read_dat_file('outPV.dat')

RI = read_dat_file('outRI.dat')
RQ = read_dat_file('outRQ.dat')
RU = read_dat_file('outRU.dat')
RV = read_dat_file('outRV.dat')

VI = read_dat_file('outVI.dat')
VQ = read_dat_file('outVQ.dat')
VU = read_dat_file('outVU.dat')
VV = read_dat_file('outVV.dat')

# Function to find the extreme value with its sign
def find_extreme_value_with_sign(matrix):
    max_value = np.max(matrix)
    min_value = np.min(matrix)
    return max_value if abs(max_value) >= abs(min_value) else min_value

# Function to find the maximum positive value
def find_max_positive_value(matrix):
    return np.max(matrix[matrix > 0]) if np.any(matrix > 0) else -np.inf

# Find the extreme values with their signs and maximum positive values for each matrix
results = {}
for name, matrix in zip(
    ['HI', 'HQ', 'HU', 'HV', 'PI', 'PQ', 'PU', 'PV', 'RI', 'RQ', 'RU', 'RV', 'VI', 'VQ', 'VU', 'VV'],
    [HI, HQ, HU, HV, PI, PQ, PU, PV, RI, RQ, RU, RV, VI, VQ, VU, VV]
):
    extreme_value = find_extreme_value_with_sign(matrix)
    max_positive_value = find_max_positive_value(matrix)
    results[name] = {
        'Extreme Value': extreme_value,
        'Max Positive Value': max_positive_value,
        'Comparison': extreme_value / max_positive_value if max_positive_value != -np.inf else 'No positive values'
    }

# Print results
for key, value in results.items():
    print(f"{key}:")
    print(f"  Extreme Value with Sign: {value['Extreme Value']}")
    print(f"  Maximum Positive Value: {value['Max Positive Value']}")
    print(f"  Extreme Value / Max Positive Value: {value['Comparison']}")
```

```python
import numpy as np
import pandas as pd

def read_dat_file(file_path):
    """Reads a .dat file and returns the data as a numpy array."""
    return np.loadtxt(file_path)

# Read all the required .dat files
HI = read_dat_file('outHI.dat')
HQ = read_dat_file('outHQ.dat')
HU = read_dat_file('outHU.dat')
HV = read_dat_file('outHV.dat')

PI = read_dat_file('outPI.dat')
PQ = read_dat_file('outPQ.dat')
PU = read_dat_file('outPU.dat')
PV = read_dat_file('outPV.dat')

RI = read_dat_file('outRI.dat')
RQ = read_dat_file('outRQ.dat')
RU = read_dat_file('outRU.dat')
RV = read_dat_file('outRV.dat')

VI = read_dat_file('outVI.dat')
VQ = read_dat_file('outVQ.dat')
VU = read_dat_file('outVU.dat')
VV = read_dat_file('outVV.dat')

# Function to find the extreme value with its sign
def find_extreme_value_with_sign(matrix):
    max_value = np.max(matrix)
    min_value = np.min(matrix)
    return max_value if abs(max_value) >= abs(min_value) else min_value

# Function to find the maximum positive value
def find_max_positive_value(matrix):
    return np.max(matrix[matrix > 0]) if np.any(matrix > 0) else -np.inf

# Prepare data for Excel output
data = []

for name, matrix in zip(
    ['HI', 'HQ', 'HU', 'HV', 'PI', 'PQ', 'PU', 'PV', 'RI', 'RQ', 'RU', 'RV', 'VI', 'VQ', 'VU', 'VV'],
    [HI, HQ, HU, HV, PI, PQ, PU, PV, RI, RQ, RU, RV, VI, VQ, VU, VV]
):
    extreme_value = find_extreme_value_with_sign(matrix)
    max_positive_value = find_max_positive_value(matrix)
    comparison = extreme_value / max_positive_value if max_positive_value != -np.inf else 'No positive values'
    data.append({
        'Matrix': name,
        'Extreme Value with Sign': extreme_value,
        'Maximum Positive Value': max_positive_value,
        'Extreme Value / Max Positive Value': comparison
    })

# Create DataFrame
df = pd.DataFrame(data)

# Save DataFrame to Excel
output_file = 'matrix_comparisons.xlsx'
df.to_excel(output_file, index=False)

print(f"Results have been saved to {output_file}")
```
