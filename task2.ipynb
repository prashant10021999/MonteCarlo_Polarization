```python
def get_array_dimensions(filename):
    with open(filename, 'r') as file:
        # Read the first line to determine the number of columns
        first_line = file.readline().strip()
        columns = len(first_line.split())
        
        # Count the number of lines to determine the number of rows
        rows = 1
        for line in file:
            rows += 1
            
    return rows, columns

# Example usage
filename = 'outHI.dat'
rows, columns = get_array_dimensions(filename)
print(f"The array dimensions are: {rows} rows x {columns} columns")
```

```python
def get_center_element(filename):
    with open(filename, 'r') as file:
        # Read all lines
        lines = file.readlines()

        # The center of a 100x100 matrix is at position (50, 50) (0-indexed, it's (49, 49))
        center_row = 1
        center_col = 1
        
        # Extract the 50th row (index 49)
        center_line = lines[center_row].strip()
        
        # Split the line into elements and get the 50th element (index 49)
        center_value = center_line.split()[center_col]
        
    return center_value

# Example usage
filename = 'outHI.dat'
center_value = get_center_element(filename)
print(f"The center element value is: {center_value}")
```

```python
import pandas as pd

def read_2d_array(filename):
    array = []
    
    with open(filename, 'r') as file:
        for line in file:
            # Split the line into individual elements and convert them to floats
            row = list(map(float, line.split()))
            array.append(row)
    
    return array

def flatten_array(array):
    flattened = []
    index = 1  # Start indexing from 1
    
    for i, row in enumerate(array):
        for j, value in enumerate(row):
            flattened.append((index, value))
            index += 1
    
    return flattened

def save_to_excel(data, output_filename):
    df = pd.DataFrame(data, columns=['Index', 'Value'])
    df.to_excel(output_filename, index=False)

# Example usage
filename = 'outHI.dat'
output_filename = 'output.xlsx'

# Read and flatten the array
array = read_2d_array(filename)
flattened_data = flatten_array(array)

# Save to Excel
save_to_excel(flattened_data, output_filename)

print(f"Data has been saved to {output_filename}")
```

```python
def read_2d_array(filename):
    array = []
    
    with open(filename, 'r') as file:
        for line in file:
            # Split the line into individual elements and convert them to floats
            row = list(map(float, line.split()))
            array.append(row)
    
    return array

def find_max_value_index(array):
    max_value = float('-inf')
    max_index = None
    flat_index = 1  # Start indexing from 1
    
    for i, row in enumerate(array):
        for j, value in enumerate(row):
            if value > max_value:
                max_value = value
                max_index = flat_index
            flat_index += 1
    
    return max_index, max_value

# Example usage
filename = 'outRQ.dat'

# Read the 2D array
array = read_2d_array(filename)

# Find the index and value of the maximum element
max_index, max_value = find_max_value_index(array)

print(f"The maximum value is {max_value} at index {max_index}")
```
